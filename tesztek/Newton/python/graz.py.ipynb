```python
# http://systems-sciences.uni-graz.at/etextbook/sw2/lyapunov.html

import numpy as np
import matplotlib.pyplot as plt
import numpy as np
# show plots in notebook
%matplotlib inline


lambdas = []
maps = []

# define range of r 
rValues = np.linspace(0.02, 2, 200)
initRange=200

# loop through r
for r in rValues:
    x = 0.1
    x= 0.5
    x=np.random.rand()
    result = 0.0
    # iterate system 100 times
    for t in range(initRange):
        x = x + r - x**2
        # calculate log of the absolute of the derivative
        result += np.log(np.abs(1 - 2*x))
    # take average
    lambdas.append(result / initRange)
    # for the map ignore first 100 iterations as transient time and iterate anew
    for t in range(20):
        x = x + r - x**2
        maps.append(x)    
    
fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(1,1,1)

xticks = np.linspace(0, 2, 4000)
# zero line
zero = [0]*4000
ax1.plot(xticks, zero, 'g-')
# plot map
ax1.plot(xticks, maps, 'r.',alpha = 0.3, label = 'Map')
ax1.set_xlabel('r')
# plot lyapunov
ax1.plot(rValues, lambdas, 'b-', linewidth = 3, label = 'Lyapunov exponent')
ax1.grid('on')
ax1.set_xlabel('r')
ax1.legend(loc='best')
ax1.set_title('Map of x(t+1) = x(t) + r - x(t)^2 versus Lyapunov exponent')
```
